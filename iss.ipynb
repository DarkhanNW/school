```python
import requests, matplotlib.pyplot as plt, time
iss_api_url = "https://api.wheretheiss.at/v1/satellites/25544"
interval_seconds = 10
longitude_points, latitude_points = [], []

def ping_iss_location(api_url):
    response = requests.get(api_url)
    response.raise_for_status()
    data = response.json()
    print(f"latitude: {data['latitude']}")
    print(f"longitude: {data['longitude']}")
    print(f"timestamp: {data['timestamp']}")
    return data["latitude"], data["longitude"], data["timestamp"]
        
for i in range(100):
    print(f"Pinging API at {time.ctime()}")
    latitude, longitude, timestamp = ping_iss_location(iss_api_url)
    latitude_points.append(latitude)
    longitude_points.append(longitude) 
    time.sleep(interval_seconds)

fig, ax = plt.subplots()
ax.set(xlabel="Longitude", ylabel="Latitude", xlim=(-180, 180), ylim=(-90, 90), title="ISS Path")
ax.plot(longitude_points, latitude_points, "o")
plt.show()
```
